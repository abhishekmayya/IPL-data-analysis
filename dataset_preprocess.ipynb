#Total runs scored and wicket taken for a player in each match

```python


# Group the DataFrame by match and batter to calculate the total runs scored by each batter in each match
batter_runs = merged_data.groupby(['ID', 'batter'])['batsman_run'].sum().reset_index()

# Group the DataFrame by match and bowler to calculate the total wickets taken by each bowler in each match
excluded_kinds = ['retired hurt', 'retired out', 'run out', 'obstructing the field']
bowler_wickets = merged_data[merged_data['isWicketDelivery'] == 1 & (~merged_data['kind'].isin(excluded_kinds))].groupby(['ID', 'bowler'])['kind'].count().reset_index()

# Rename the columns for clarity
batter_runs.rename(columns={'batsman_run': 'TotalRunsScored'}, inplace=True)
bowler_wickets.rename(columns={'kind': 'WicketsTaken'}, inplace=True)

batter_runs.sort_values(by=['TotalRunsScored'], ascending=False, inplace=True)
bowler_wickets.sort_values(by=['WicketsTaken'], ascending=False, inplace=True)
```

```python

print("Runs Scored by Each Batter in Each Match:")
print(batter_runs)
```

```python
print("\nWickets Taken by Each Bowler in Each Match:")
print(bowler_wickets)
```

#Total runs scored and wicket taken for a player in each season

```python
season_run_getters = merged_data.groupby(['Season', 'batter'])['batsman_run'].sum().reset_index()

# Group the DataFrame by Season and bowler to calculate total wickets taken by each bowler in each season
excluded_kinds = ['retired hurt', 'retired out', 'run out', 'obstructing the field']
season_wicket_takers = merged_data[merged_data['isWicketDelivery'] == 1 & (~merged_data['kind'].isin(excluded_kinds))].groupby(['Season', 'bowler'])['kind'].count().reset_index()

# Rename the columns for clarity
season_run_getters.rename(columns={'batsman_run': 'TotalRunsScored'}, inplace=True)
season_wicket_takers.rename(columns={'kind': 'TotalWicketsTaken'}, inplace=True)
season_run_getters.sort_values(by=['TotalRunsScored'], ascending=False, inplace=True)
season_wicket_takers.sort_values(by=['TotalWicketsTaken'], ascending=False, inplace=True)
```

```python
print("Run Getters for Each Season:")
print(season_run_getters)
```
